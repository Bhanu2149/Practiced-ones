```python
ages= np.array([22,23,25,27,30])
salaries= np.array([30000,28000,29000,25000,29500])

data = {

    "Name" : ["Bhanu", "Ramu", "Srinu", "sai", "Madhu"],
    "Age" : ages,
    "Salary" : salaries,
    "Country" : ["India", "Chicago", "Usa", "NY", "La"]
}

df= pd.DataFrame(data)

df["Bonus"] = df["Salary"] * 0.2

df["Country"] = df["Country"].replace("India", "Bharat")

print(df)

print("\nName & Salary :")
print(df[["Name", "Salary"]])

print("\nBelow 26 age people details :")
print(df[df["Age"] < 26])

print("\nGroup :")
print(df.groupby("Country")["Salary"].mean())

print("\nSorting: ")
print(df.sort_values("Salary", ascending=False))

print("\nAvg Salary:")
print(df.loc[df["Salary"].idxmax()])

print("\nAvarage:")
print(df["Salary"].mean())
```

```python
np.random.seed(42)

students = ["S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "S10"]
maths = np.random.randint(50, 100, size=10)
science = np.random.randint(40, 95, size=10)
english = np.random.randint(45, 90, size=10)
history = np.random.randint(30, 85, size=10)

df = pd.DataFrame({
    "Student": students,
    "Maths": maths,
    "Science": science,
    "English": english,
    "History": history
})

print(df)

print("\nBasic Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nAverage Marks:")
print(df.mean(numeric_only=True))

avg_marks = df.mean(numeric_only=True)

plt.bar(avg_marks.index, avg_marks.values, color=["blue", "green", "orange", "red"])
plt.title("Average Marks per Subject")
plt.ylabel("Marks")
plt.show()

plt.figure(figsize=(10,5))
sns.heatmap(df.drop(columns="Student").corr(), annot=True, cmap="coolwarm")
plt.title("Subject Correlation Heatmap")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=df.drop(columns="Student"))
plt.title("Marks Distribution by Subject")
plt.show()

sns.pairplot(df.drop(columns="Student"))
plt.show()


```
